```python
import os, re
from collections import defaultdict

docs = {
    "doc1": "Information Retrieval focuses on searching and indexing.",
    "doc2": "Machine learning supports intelligent information systems.",
    "doc3": "Retrieval models use Boolean and probabilistic methods.",
}

dictionary = set()
for content in docs.values():
    tokens = re.findall(r'\b\w+\b', content.lower())
    dictionary.update(tokens)

print(dictionary)
```

```python
inverted_index = defaultdict(list)

for doc_id, content in docs.items():
    tokens = re.findall(r'\b\w+\b', content.lower())
    for token in set(tokens):
        inverted_index[token].append(doc_id)

print(inverted_index)
```

```python
def boolean_and(term1, term2):
    return set(inverted_index[term1]) & set(inverted_index[term2])

def boolean_or(term1, term2):
    return set(inverted_index[term1]) | set(inverted_index[term2])

def boolean_not(term):
    return set(docs.keys()) - set(inverted_index[term])
```

```python
print(boolean_and("retrieval", "information"))
```
